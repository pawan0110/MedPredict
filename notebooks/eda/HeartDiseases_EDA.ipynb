```python
import pandas as pd
import numpy as np

from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("../../datasets/synthetic_heart_disease_dataset.csv")
df.head()
```

```python
print("Shape of Dataset:", df.shape)
print("\nData Types:\n")
print(df.dtypes)
```

```python
df['Alcohol_Intake'].value_counts()
```

```python
sns.countplot(x='Heart_Disease', data=df)
plt.title("Heart Disease Distribution")
plt.show()

df['Heart_Disease'].value_counts()
```

```python
numerical_cols = df.select_dtypes(include=np.number).columns

df[numerical_cols].hist(bins=20, figsize=(15,12))
plt.tight_layout()
plt.show()
```

```python
categorical_cols = df.select_dtypes(include=['object', 'string']).columns

for col in categorical_cols:
    plt.figure()
    sns.countplot(x=col, data=df)
    plt.xticks(rotation=45)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
sns.boxplot(x='Heart_Disease', y='Age', data=df)
plt.title("Age vs Heart Disease")
plt.show()
```

```python
sns.boxplot(x='Heart_Disease', y='BMI', data=df)
plt.title("BMI vs Heart Disease")
plt.show()
```

```python
sns.boxplot(x='Heart_Disease', y='Systolic_BP', data=df)
plt.show()

sns.boxplot(x='Heart_Disease', y='Diastolic_BP', data=df)
plt.show()
```

```python
for col in categorical_cols:
    plt.figure()
    sns.countplot(x=col, hue='Heart_Disease', data=df)
    plt.xticks(rotation=45)
    plt.title(f"{col} vs Heart Disease")
    plt.show()
```

```python
plt.figure(figsize=(12,8))

numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(numeric_df.corr(), annot=False, cmap='coolwarm')

plt.title("Correlation Matrix")
plt.show()
```

```python
corr = df.corr(numeric_only=True)['Heart_Disease'].sort_values(ascending=False)
print(corr)
```

```python
for col in numerical_cols:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f"Outliers in {col}")
    plt.show()
```

```python
sns.pairplot(df[['Age','BMI','Systolic_BP','Cholesterol_Total','Heart_Disease']],
             hue='Heart_Disease')
plt.show()
```
